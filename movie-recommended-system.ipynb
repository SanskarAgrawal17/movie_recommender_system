```python
import ast
def find_value(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(find_value)
```

```python
import ast
def find_value(text):
    l=[]
    count=0
    for i in ast.literal_eval(text):
        if count != 5 :
            l.append(i['name'])
            count+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(find_value)
```

```python
movies['cast'].head(2)
```

```python
import ast
def find_value1(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director' :
            l.append(i['name'])
            break
        
    return l
```

```python
movies['crew']=movies['crew'].apply(find_value1)
```

```python
m=[]
for i in new_data['tags']:
  m.append(i.lower())
```

```python
new_data['tags']=m
```

```python
new_data['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector= cv.fit_transform(new_data['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
y=[]
def stem(text):
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_data['tags']=new_data['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise  import cosine_similarity
cosine_similarity(vector)
```

```python
similarity=cosine_similarity(vector)
```

```python
new_data['title'].head(20)
```

```python
def recommend(movie):
    movie_index= new_data[new_data['title'] == movie].index[0]
    distance=similarity[movie_index]
    movie_list= sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
